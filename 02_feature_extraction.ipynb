```python
import re
def rep_tut(replie):
    # парсим ответы. Количество и есть ли ответ препода
    replies = 0
    tutor = False
    if type(replie) == str:
        p = re.compile('.*([0-9]+).*')
        m = p.match(replie.replace("\n", " "))
        if m:
            replies = int(m.groups()[0])
        p = re.compile(".*(ответ преподавателя).*")
        if p.match(replie.replace("\n", " ")):
            tutor = True
            replies -= 1
    return (replies, tutor)

def rep1(replie):
    return rep_tut(replie)[0]

def rep2(replie):
    return rep_tut(replie)[1]

```

```python
comments.loc[:, 'body_length'] = comments.body.apply(lambda b: len(b))
comments.loc[:, 'body_question'] = comments.body.apply(lambda b: "?" in b).astype(int)
comments.loc[:, 'body_tags'] = comments.body.apply(lambda b: "img" in b or "li" in b or "href" in b).astype(int)
```

```python
def resh(body):
    p = re.compile('.*([0-9][0-9][0-9][0-9][0-9]+).*')
    m = p.match(body.replace("\n", " "))
    return bool(m) or 'реше' in body
comments.loc[:, 'body_solu'] = comments.body.apply(resh).astype(int)
```

```python
authors = comments.groupby('author')
aut_dreq = defaultdict(lambda x: 0)
for key, a in authors:
    times = a.time.apply(lambda t: datetime.strptime(str(t), '%Y-%m-%dT%H:%M:%S'))
    aut_dreq[key] = len(a)/((max(times) - min(times)).total_seconds()/60.0/60/24) if (len(a)-1) else 0
comments.loc[:, 'user_freq'] = comments.author.apply(aut_dreq.get)
m = comments.user_freq.mean()
comments.loc[:, 'user_freq'] = comments.user_freq.apply(lambda uf: uf if uf else m)
```

```python
comments.loc[:, 'step_type'] = comments.step_link.apply(lambda link: steps[steps.step_link==link].s_type.values[0])
steps_ = comments.groupby('step_link')
step_dreq = defaultdict(lambda x: 0)
for key, a in steps_:
    times = a.time.apply(lambda t: datetime.strptime(str(t), '%Y-%m-%dT%H:%M:%S'))
    step_dreq[key] = len(a)/((max(times) - min(times)).total_seconds()/60.0/60/24) if (len(a)-1) else 0
comments.loc[:, 'step_freq'] = comments.step_link.apply(step_dreq.get)
m = comments.user_freq.mean()
comments.loc[:, 'step_freq'] = comments.step_freq.apply(lambda uf: uf if uf else m)
```

```python
comments.loc[:, 'reply_num'] = comments.replies.apply(rep1)
comments.loc[:, 'tutor_mark'] = comments.replies.apply(rep2).astype(int);
del comments['replies']
```
