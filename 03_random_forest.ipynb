```python
import pandas as pd
import numpy as np
from collections import defaultdict
from sklearn.ensemble import RandomForestClassifier
from sklearn.cross_validation import train_test_split
from sklearn.metrics import roc_auc_score
```

```python
data = pd.DataFrame.from_csv('res.csv')
test = pd.DataFrame.from_csv('test.csv')
```

```python
data
```

```python
Y = data.tutor_mark.values.astype(int)
```

```python
types = pd.get_dummies(data.step_type)
```

```python
X = data.loc[:, ['body_question', 'body_length', 'body_tags', 'body_solu','user_freq', 'step_freq', 'reply_num']].join(types)
```

```python
X
```

```python
model =  RandomForestClassifier(class_weight = 'balanced', n_estimators=100, max_depth=4)
#Поделим нашу выборку на train и validation
Xtr,Xval,Ytr,Yval = train_test_split(X,Y,test_size=0.10,random_state=128)
```

```python
model.fit(Xtr, Ytr)

#Оценим метрику на validation
print(roc_auc_score(Yval, model.predict(Xval)))
print(roc_auc_score(Ytr, model.predict(Xtr)))
```

```python
types = pd.get_dummies(test.step_type)
X_test = test.loc[:, ['body_question', 'body_length', 'body_tags','body_solu','user_freq', 'step_freq', 'reply_num']].join(types)
```

```python
test.loc[:, 'result'] = model.predict_proba(X_test)[:, 0]
```

```python
test.sort_values(by='result').to_csv('render_ready.csv')
```

```python
Y_test = test.tutor_mark.values.astype(int)
```

```python
print(roc_auc_score(Y_test, model.predict(X_test)))
```
