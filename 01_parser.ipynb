```python
from bs4 import BeautifulSoup
import numpy as np
import html
import collections
```

```python
# комментарии парсим
with open ("data_grabber/data/gamefy_end.txt", "r") as myfile:
    data=myfile.readlines()
```

```python
content = "\n".join(data)
```

```python
soup = BeautifulSoup(content, "html.parser")
```

```python
comments = soup.find_all('div', {'class': "comment-widget"})
```

```python
Half_RAW_data = collections.namedtuple('Half_RAW_data', ['c_link', 'author', 'step_link', 'body', 'replies', 'time'])
```

```python
half_RAW_data = []
for comment in comments:
    author = int(str(comment.find('span', {'class': 'comment-widget__avatar'}).find('a').attrs['href']).split('/')[-1])
    c_link = comment.find('a', {'class': 'comment-widget__link'}).attrs['href']
    body = "\n".join(map(str,comment.find('div', {'class': 'comment-widget__body'}).contents))
    body = body.replace('\xa0', ' ').replace('\ufeff', '').replace('\n', " ").replace("\t", " ").strip()
    step_link = comment.find('a', {'class': 'comment-widget__target-link'}).attrs['href'].split("?")[0]
    replies = comment.find('span', {'class': 'comment-widget__replies-counter'})
    time = comment.find('time').attrs['datetime']
    half_RAW_data.append(Half_RAW_data(author=author, c_link=c_link, step_link=step_link, body=body, replies=replies, time=time))
```

```python
# степы парсим

Half_RAW_step = collections.namedtuple('Half_RAW_step', ['step_link', 's_type', 'body', 'task', 'quize_typename'])
```

```python
with open ("data_grabber/data/struc_end.txt", "r") as myfile:
    data=myfile.readlines()
content = "\n".join(data)
soup = BeautifulSoup(content, "html.parser")
players = soup.find_all('div', {'class': "lesson__player"})
```

```python
half_raw_steps = []
for player in players:
    step_type = " ".join(player.attrs['class']).strip().split()[-1]
    step_body = "\n".join(map(str, player.find('div', {'class': 'step-inner'}).contents))
    step_body = step_body.replace('\xa0', ' ').replace('\ufeff', '').replace('\n', " ").replace("\t", " ").strip()
    step_task = player.find('div', {'class': 'attempt'})
    step_quiz_typename = player.find('div', {'class': 'quiz__typename'})
    step_link = player.find('a', {'class': 'active'}).attrs['href'].split('?')[0]
    half_raw_steps.append(Half_RAW_step(step_link=step_link, s_type=step_type, body=step_body, task=step_task, quize_typename=step_quiz_typename))
```

```python
import pandas as pd
```

```python
comments = pd.DataFrame.from_records(half_RAW_data, columns=Half_RAW_data._fields)
comments.to_csv('comments.csv')
comments
```

```python
steps = pd.DataFrame.from_records(half_raw_steps,columns=Half_RAW_step._fields)
steps.to_csv('steps.csv')
steps
```
